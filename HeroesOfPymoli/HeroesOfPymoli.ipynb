```python
#Pandas library
import pandas as pd
```

```python
#Load csv file
game_data = pd.read_csv("purchase_data.csv")
```

```python
#review csv file
game_data.head()
```

```python
#Review total number of players
num_players = game_data["SN"].nunique()
num_items = game_data["Item ID"].nunique()
avg_purch = game_data["Price"].mean()
num_purch = game_data["Purchase ID"].count()
tot_revenue = game_data["Price"].sum()
```

```python
#Purchasing Analysis (Total)
purch_df = pd.DataFrame({"Total_Items":[num_items],
                         "Avg_Price":[avg_purch],
                         "Total_Purchases":[num_purch],
                         "Total_Revenue":[tot_revenue],
                         "Total_Players":[num_players]})
purch_df['Avg_Price'] = purch_df['Avg_Price'].map('${:,.2f}'.format)
purch_df['Total_Revenue'] = purch_df['Total_Revenue'].map('${:,.2f}'.format)
purch_df
```

```python
#Demographics Analysis (Gender)
gender = game_data.groupby(["Gender"])
pergen = (gender["SN"].nunique()/num_players)
countgen = gender["SN"].nunique()
stat_gen = pd.DataFrame({"Total_Players": countgen,
                         "%_Players": pergen})
stat_gen['%_Players'] = stat_gen['%_Players'].map('{:.2%}'.format)
stat_gen
```

```python
#Purchasing Analysis (Gender) 
purchgen = gender["Purchase ID"].count()
sumpurchgen = gender["Price"].sum()
avgpurch = sumpurchgen / purchgen
avgpurchper = sumpurchgen / countgen
purch_gen = pd.DataFrame({"Players_Count":purchgen,
                          "Avg_Purchase":avgpurch,
                          "Total_Purchase":sumpurchgen,
                          "Avg_Total_Purchase_Person":avgpurchper})
purch_gen['Avg_Purchase'] = purch_gen['Avg_Purchase'].map('${:,.2f}'.format)
purch_gen['Total_Purchase'] = purch_gen['Total_Purchase'].map('${:,.2f}'.format)
purch_gen['Avg_Total_Purchase_Person'] = purch_gen['Avg_Total_Purchase_Person'].map('${:,.2f}'.format)
purch_gen
```

```python
#Demographics Analysis (Age)
age_lbl = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]
cut_age = [0,9,14,19,24,29,34,39,100]
game_data["Range_age"] = pd.cut(game_data["Age"], bins=cut_age, labels=age_lbl)
age_data_df = game_data.groupby(["Range_age"])
countage = age_data_df["SN"].nunique() 
perage = (age_data_df["SN"].nunique()/num_players)
stat_age = pd.DataFrame({"Total_Players": countage,
                         "%_Players": perage})
stat_age['%_Players'] = stat_age['%_Players'].map('{:.2%}'.format)
stat_age
```

```python
purchage = age_data_df["Purchase ID"].count()
sumpurchage = age_data_df["Price"].sum()
avgpurchage = sumpurchage / purchage
avgpurchageper = sumpurchage / countage
purch_age = pd.DataFrame({"Players_Count":purchage,
                          "Avg_Purchase":avgpurchage,
                          "Total_Purchase":sumpurchage,
                          "Avg_Total_Purchase_Person":avgpurchageper})
purch_age['Avg_Purchase'] = purch_age['Avg_Purchase'].map('${:,.2f}'.format)
purch_age['Total_Purchase'] = purch_age['Total_Purchase'].map('${:,.2f}'.format)
purch_age['Avg_Total_Purchase_Person'] = purch_age['Avg_Total_Purchase_Person'].map('${:,.2f}'.format)
purch_age
```

```python
player_df = game_data.groupby(["SN"])
countplay = player_df["Purchase ID"].count()
totpurchplay = player_df["Price"].sum()
avgpurchplay = totpurchplay / countplay
stat_play = pd.DataFrame({"Count": countplay,
                          "Avg_Purchase_Price":avgpurchplay,
                          "Total_Purchase_Val":totpurchplay})
stat_play['Avg_Purchase_Price'] = stat_play['Avg_Purchase_Price'].map('${:,.2f}'.format)
stat_play['Total_Purchase_Val'] = stat_play['Total_Purchase_Val'].map('${:,.2f}'.format)
stat_play.sort_values(by='Total_Purchase_Val',ascending=False)
stat_play.head()
```

```python
item_df = game_data.groupby(["Item Name"])
countitem = item_df["Purchase ID"].count()
totpurchitem = item_df["Price"].sum()
itemprice = item_df["Price"].value()
stat_item = pd.DataFrame({"Count": countitem,
                          "Item_Price":itemprice,
                          "Total_Purchase_Val":totpurchitem})
stat_item
```
